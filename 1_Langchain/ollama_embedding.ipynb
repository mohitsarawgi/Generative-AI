# Ollama
- Ollama supports embedding models,
- making it possible to build retrival augmented generation (RAG) applications
- that combines text promts with existing documents or other data.

- Ollama is an open-source platform designed to make it easy to run large language models (LLMs) locally on your own computer

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")  ## by default it uses llama2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek Alphabet",
        "Beta is the second letter of Greek Alphabet",

    ]
)
```

```python
r1
```

```python
# dimentions = 2048
len(r1[0])
```

```python
embeddings.embed_query("What is the second letter of greek alphabet")
```
